## LDNe (do not run) - sample size effects

```python
m = Bottleneck(gens=20)
m.bgen = 10
m.num_msats = 100
m.start_size = 200
m.end_size = 50
m.sample_size = [10, 20, 50]
BasicView(m, [LDNe()], max_y=[m.start_size * 3])
m.run()
```
